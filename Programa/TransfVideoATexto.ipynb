```python
import tkinter as tk
from tkinter import filedialog
import wave
import math
import contextlib
import speech_recognition as sr
from moviepy.editor import AudioFileClip

def buscar_archivo():
    global zoom_video_file_name
    zoom_video_file_name = filedialog.askopenfilename(filetypes=[("Archivos de video", "*.mp4;*.avi;*.mkv"), ("Todos los archivos", "*.*")])
    if zoom_video_file_name:
        label_archivo.config(text=f"Archivo seleccionado: {zoom_video_file_name}")

def transcribir_video():
    if not zoom_video_file_name:
        label_archivo.config(text="¡Selecciona un archivo de video primero!")
        return

    transcribed_audio_file_name = "transcribed_speech.wav"

    audioclip = AudioFileClip(zoom_video_file_name)
    audioclip.write_audiofile(transcribed_audio_file_name)

    with contextlib.closing(wave.open(transcribed_audio_file_name,'r')) as f:
        frames = f.getnframes()
        rate = f.getframerate()
        duration = frames / float(rate)

    total_duration = math.ceil(duration / 60)

    r = sr.Recognizer()

    transcription_result = ""
    for i in range(0, total_duration):
        with sr.AudioFile(transcribed_audio_file_name) as source:
            audio = r.record(source, offset=i*60, duration=60)
        transcription_result += r.recognize_google(audio, language='es-ES') + " "

    text_transcripcion.config(state=tk.NORMAL)
    text_transcripcion.delete(1.0, tk.END)
    text_transcripcion.insert(tk.END, transcription_result)
    text_transcripcion.config(state=tk.DISABLED)

    label_archivo.config(text="Transcripción completada. El resultado se muestra abajo.")

# Crear la ventana principal
ventana = tk.Tk()
ventana.title("Transcripción de Archivos de Video")

# Etiqueta
etiqueta = tk.Label(ventana, text="Transcribir archivos de video a texto")
etiqueta.pack(pady=10)

# Botón para buscar archivo de video
boton_buscar = tk.Button(ventana, text="Buscar Archivo de Video", command=buscar_archivo)
boton_buscar.pack(pady=5)

# Etiqueta para mostrar el archivo seleccionado
label_archivo = tk.Label(ventana, text="Archivo seleccionado: Ninguno")
label_archivo.pack(pady=5)

# Cuadro de texto para mostrar el resultado de la transcripción
text_transcripcion = tk.Text(ventana, height=10, width=50, state=tk.DISABLED)
text_transcripcion.pack(pady=10)

# Botón para transcribir video
boton_transcribir = tk.Button(ventana, text="Transcribir Video", command=transcribir_video)
boton_transcribir.pack(pady=10)

# Iniciar el bucle de la interfaz gráfica
ventana.mainloop()

```
